```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def newt(f, df, x0, nmax):
  xs = []
  x = x0
  xs.append(x)
  for _ in range(nmax):
    x = x - f(x) / df(x)
    xs.append(x)
  return np.array(xs)
```

```python
f = lambda x: ((x - 2) * x + 1) * x - 3
df = lambda x: (3 * x - 4) * x + 1
x0 = 3
nmax = 5
root = newt(f, df, x0, nmax)[-1]
assert np.isclose(f(root), 0)
```

```python
xs = np.linspace(0, 3.5, 1000)
ys = f(xs)
plt.plot(xs, ys, 'b--')
plt.plot([root], [f(root)], 'ro')
plt.plot(xs, np.zeros_like(xs), 'r--')
```
